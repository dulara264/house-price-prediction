```python
plt.figure(figsize=(6,4))
sns.histplot(df["SalePrice"], kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("House Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["SalePrice"], kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("House Price")
plt.ylabel("Count")
plt.savefig("../plots/price_distribution.png", bbox_inches="tight")
plt.show()
```
